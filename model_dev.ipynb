### Exploratory Analysis of Stock Data

```python
# Obtain historical data of a stock 
ticker = 'HLT'
start = '1997-01-01'
end = '2022-08-30'

yf_obj = yf.Ticker(ticker)
df = yf_obj.history(start=start, end=end)
```

### Exploratory Analysis of CPI and Interest Rate

```python
df_cpi = pd.read_csv("data/CPIAUCSL.csv")
df_funds = pd.read_csv("data/FEDFUNDS.csv")
```

```python
df_cpi.head()
```

```python
def append_columns(df, df_extra_col):
    df_extra_col['DATE'] = df_extra_col['DATE'].astype('datetime64[ns]')
    df_extra_col = df_extra_col.set_index('DATE')
    df_merged = df.merge(df_extra_col, left_index=True, right_index=True, how='left')
    return df_merged

data = df
data = append_columns(data, df_cpi)
data = append_columns(data, df_funds)
data.head()
```

```python
data.info()
```

```python
# Fill in the first entry for CPI and funds rate
data.CPIAUCSL[0] = 159.4
data.FEDFUNDS[0] = 5.25
data = data.fillna(method='ffill', axis=0)
```

```python
data.to_csv(f"data/{ticker}_raw_data.csv")
```
