```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# Obtain historical data of a stock 
ticker = 'AAPL'
start = '1997-01-01'
end = '2022-08-30'

yf_obj = yf.Ticker(ticker)
data = yf_obj.history(start=start, end=end)
```
